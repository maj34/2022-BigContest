```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

# Data Load

```python
master_train_oversample = pd.read_csv('../Data/master_train_oversample.csv') ; master_train_oversample.head()
master_public_data = pd.read_csv('../Data/master_public_data.csv') ; master_public_data.head()
```

```python
master_public_label = master_public_data['is_applied']
master_public_data = master_public_data.drop('is_applied', axis=1)
```

```python
# https://pycaret.readthedocs.io/en/stable/api/classification.html

#cat_feats = ['income_type', 'employment_type', 'houseown_type', 'purpose', 'personal_rehabilitation_yn', 'personal_rehabilitation_complete_yn', 
#             'gender', 'bank_diversity', 'product_diversity', 'holiday', 'weekday', 'foreign', 'segment', 'bank_label', 'product_label'] categorical_features=cat_feats

exp_clf = setup(data = master_train_oversample, target = 'is_applied', train_size = 0.8, data_split_shuffle=True,
                session_id=123, use_gpu = True, test_data = master_public_data)
```

```python
best_model = compare_models(sort="F1", fold=3)
```

```python
best_model
```

```python
save_model(best_model, '../Model/pycaret_et')
```

```python
prediction = predict_model(best_model, data = master_public_data)
```
