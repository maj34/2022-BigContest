# Data Merge

```python
train = pd.merge(x_train,bank_df,on='bank_id',how='left')
public = pd.merge(x_public,bank_df,on='bank_id',how='left')
private = pd.merge(x_private,bank_df,on='bank_id',how='left')
test = pd.merge(test,bank_df,on='bank_id',how='left')
```

```python
train = pd.merge(train,product_df,on='product_id',how='left')
public = pd.merge(public,product_df,on='product_id',how='left')
private = pd.merge(private,product_df,on='product_id',how='left')
test = pd.merge(test,product_df,on='product_id',how='left')
```

```python
target = train['is_applied']
train.drop(['is_applied'],axis=1,inplace=True)
train['is_applied'] = target
```

```python
target = public['is_applied']
public.drop(['is_applied'],axis=1,inplace=True)
public['is_applied'] = target
```

```python
target = private['is_applied']
private.drop(['is_applied'],axis=1,inplace=True)
private['is_applied'] = target
```

```python
target = test['is_applied']
test.drop(['is_applied'],axis=1,inplace=True)
test['is_applied'] = target
```

# Data Drop

```python
submission
```

```python
drop_columns = ['application_id','bank_id','product_id']
```

```python
train.drop(drop_columns,axis=1,inplace=True)
public.drop(drop_columns,axis=1,inplace=True)
private.drop(drop_columns,axis=1,inplace=True)
#test.drop(drop_columns,axis=1,inplace=True)
```

# LabelEncoding

```python
cat_feature = ['income_type','employment_type','houseown_type']
```

```python
le = LabelEncoder()
for i in cat_feature:
    train[i] = le.fit_transform(train[i])
    public[i] = le.transform(public[i])
    private[i] = le.transform(private[i])
    test[i] = le.transform(test[i])
```
