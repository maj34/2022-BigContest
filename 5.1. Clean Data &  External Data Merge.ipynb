## 외부데이터 병합 

```python
# price_index 병합
master_df = pd.merge(master_df, price, left_on='month',right_on = 'Date', how = 'left').drop('Date', axis = 1)


# lending 정보 결합
master_df = pd.merge(master_df, k_lending, left_on='month_day', right_on ='Date',how = 'left').drop('Date', axis = 1)
master_df = pd.merge(master_df, us_lending, left_on='month_day', right_on ='Date',how = 'left').drop('Date', axis = 1)

# coin 정보 결합
master_df = pd.merge(master_df, coin, left_on='month_day', right_on ='date',how = 'left').drop('date', axis = 1)

# kospi 정보 결합
master_df = pd.merge(master_df, kospi, left_on='month_day', right_on ='Date',how = 'left').drop('Date', axis = 1)
```

```python
tp_master = master_df

# 나이컬럼 만들기
tp_master['나이'] = 2022 - tp_master.birth_year + 1

# 함수 생성
def age_cat(x):
    if x < 19:
        return '15 - 19세'
    elif x >= 20 and x < 30:
        return '20 - 29세'
    elif x >= 30 and x < 40:
        return '30 - 39세'
    elif x >= 40 and x < 50:
        return '40 - 49세'
    elif x >= 50 and x < 60:
        return '50 - 59세'
    elif x >= 60:
        return '60세이상'

# 연령대로 변환하기
tp_master['연령계층별'] = tp_master['나이'].apply(age_cat)
```

```python
## 날짜 / 성별 / 나이로 DF들 병합하기
# 실업자 수 병합
tp_master = pd.merge(tp_master, unemployment_count, left_on=['month', 'gender', '연령계층별'], right_on = ['Date', '성별', '연령계층별'], how = 'left').drop(['Date', '성별'], axis = 1)

# 실업률 병합
tp_master = pd.merge(tp_master, unemployment_rate, left_on=['month', 'gender', '연령계층별'], right_on = ['Date', '성별', '연령계층별'], how = 'left').drop(['Date', '성별'], axis = 1)
```

# csi지수 병합

```python
def csi_age_cat(x):
    if x < 40:
        return '40세미만'
    elif x >= 40 and x < 50:
        return '40-50세'
    elif x >= 50 and x < 60:
        return '50-60세'
    elif x >= 60 and x < 70:
        return '60-70세'
    elif x >= 70:
        return '70세이상'
    
def csi_income_cat(x):
    if x < 1000000:
        return '100만원미만'
    elif x >= 1000000 and x < 2000000:
        return '100-200만원'
    elif x >= 2000000 and x < 3000000:
        return '200-300만원'
    elif x >= 3000000 and x < 4000000:
        return '300-400만원'
    elif x >= 4000000 and x < 5000000:
        return '400-500만원'
    elif x >= 5000000:
        return '500만원이상'
```

```python
# 병합을 위한 열 생성

tp_master['연령대'] = tp_master['나이'].apply(csi_age_cat)
tp_master['소득수준'] = np.round(tp_master['yearly_income'] / 12).apply(csi_income_cat)
```

```python
## csi index와 tp_master frame 병합
for csi_col in csi_index.CSI코드.unique():
    
    ### 현재생활형편CSI
    # 병합 임시 DF_성별
    tp_생활_성별 = csi_index[(csi_index.CSI코드 == f'{csi_col}') & ((csi_index.구분기준 == '남자') | (csi_index.구분기준 == '여자'))]

    tp_생활_성별 = tp_생활_성별.rename(columns = {'score': f'{csi_col}_성별',
                              '구분기준': 'gender'}).drop('CSI코드', axis = 1)

    tp_생활_성별.gender = tp_생활_성별.gender.apply(lambda x : 1 if x == '남자' else 0)
    tp_생활_성별.Date = tp_생활_성별.Date.apply(lambda x: x[:7])

    
    ## 나이별
    tp_생활_나이별 = csi_index[(csi_index.CSI코드 == f'{csi_col}') & ((csi_index.구분기준 == '40세미만') | (csi_index.구분기준 == '40-50세') \
                                                  | (csi_index.구분기준 == '50-60세') | (csi_index.구분기준 == '60-70세') | (csi_index.구분기준 == '70세이상'))]

    tp_생활_나이별 = tp_생활_나이별.rename(columns = {'score': f'{csi_col}_나이',
                              '구분기준': '연령대'}).drop('CSI코드', axis = 1)
    tp_생활_나이별.Date = tp_생활_나이별.Date.apply(lambda x: x[:7])

    
    ## 소득별 
    tp_생활_소득별 = csi_index[(csi_index.CSI코드 == f'{csi_col}') & ((csi_index.구분기준 == '100만원미만') | (csi_index.구분기준 == '100-200만원') | (csi_index.구분기준 == '200-300만원')\
                                                   | (csi_index.구분기준 == '300-400만원') | (csi_index.구분기준 == '400-500만원')| (csi_index.구분기준 == '500만원이상'))]

    tp_생활_소득별 = tp_생활_소득별.rename(columns = {'score': f'{csi_col}_소득',
                                            '구분기준': '소득수준'}).drop('CSI코드', axis = 1)
    tp_생활_소득별.Date = tp_생활_소득별.Date.apply(lambda x: x[:7])


    tp_master = pd.merge(tp_master, tp_생활_성별, left_on =['month','gender'], right_on = ['Date', 'gender'], how = 'left').drop('Date',axis = 1)
    tp_master = pd.merge(tp_master, tp_생활_나이별, left_on =['month','연령대'], right_on = ['Date', '연령대'], how = 'left').drop('Date',axis = 1)
    tp_master = pd.merge(tp_master, tp_생활_소득별, left_on =['month','소득수준'], right_on = ['Date', '소득수준'], how = 'left').drop('Date',axis = 1)

    
    # 해당 유저의 항목별 CSI
    tp_master[f'{csi_col}'] = np.round((tp_master[f'{csi_col}_성별'] +  tp_master[f'{csi_col}_나이'] + tp_master[f'{csi_col}_소득']) / 3)
```

```python
# 소비심리지수 생성
tp_master['소비심리지수'] = np.round((tp_master['현재생활형편CSI'] + tp_master['현재경기판단CSI'] +  tp_master['생활형편전망CSI'] +\
tp_master['소비지출전망CSI'] + tp_master['가계수입전망CSI'] + tp_master['향후경기전망CSI']) / 6, 2)
```
